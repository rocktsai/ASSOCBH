```python
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# 讀取各罪判決書語料庫
text = open("D:/期末專題/AI法官/rawData/crimeCategory/crimeCategory_13_無罪_jeba.txt",encoding="utf-8").read()
```

```python
# the font from github: https://github.com/adobe-fonts

crime = ['傷害直系血親尊親屬致死', 
         '傷害直系血親尊親屬', 
         '傷害致死', 
         '過失傷害致重傷', 
         '傷害致重傷', 
         '重傷害未遂', 
         '重傷致死', 
         '重傷害',
         '教唆傷害', # 其他
         '肇事逃逸', # 其他
         '過失傷害',
         '傷害',
         '無罪']  # 12
crimeName = crime[12]

# 讀取詞雲圖片
import imageio
back_color = imageio.v2.imread('D:/期末專題/AI法官/gavel1.png')

# 設定詞雲中文字字體 -> 需下載及設定字體，否則無法顯示中文
font = r'C:/Windows/Fonts/msjh.ttc'
wc = WordCloud(collocations=False,
               width=800,
               height=800,
               font_path=font, 
               margin=2, 
               mask=back_color, 
               background_color='white',
               #                contour_width=0.5,
               contour_color='black',
               max_words = 250
              ).generate(text)


plt.imshow(wc)
plt.axis("off")
plt.show()

wc.to_file(f'gavel_{crimeName}__.png')  # 把詞雲保存下来 
```
